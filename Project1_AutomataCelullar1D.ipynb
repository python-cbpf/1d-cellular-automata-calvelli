```python
import numpy as np
import random as rnd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
%matplotlib inline
```

```python
def concat(a, b, c):
    '''Creates the binary equivalent to the input.
    Which means it concatenates the numbers into one int
    '''
    return int(f'{a}{b}{c}')


def Automata(L,rule,time):
    '''Wolfram's Automata rule generator. 
    Takes a binary number and turns it into a rule to change states based on first neighbours    
    '''
    Matrix=[0]*L   #Stores the system
    entry=[0]*L    #Stores the states in order to synchronously update it
    
    axis=[i for i in range(L)]
    
    #000,001,010,100,011,101,110,111
    
    #Generate the initial state with 50/50 for 1 and 0
    for i in range(L):
        if(rnd.uniform(0,1.0)<= 0.5):
            Matrix[i]=1
    
    
    for time in range(time):
        #Periodic Boundary Conditions
        entry[0]=concat(Matrix[L-1], Matrix[0],Matrix[1])
        entry[L-1]=concat(Matrix[L-2], Matrix[L-1],Matrix[0])

        for i in range(1,L-1,1):
            entry[i]=concat(Matrix[i-1], Matrix[i],Matrix[i+1])
        #Applies the rule:
        for i in range(L):
            if(entry[i]==0):
                Matrix[i]=rule[0]
            elif(entry[i]==1):   #001
                Matrix[i]=rule[1]
            elif(entry[i]==10):  #010
                Matrix[i]=rule[2]
            elif(entry[i]==100):   #100
                Matrix[i]=rule[3]
            elif(entry[i]==11):   #011
                Matrix[i]=rule[4]
            elif(entry[i]==101):   #101
                Matrix[i]=rule[5]
            elif(entry[i]==110):   #110
                Matrix[i]=rule[6]
            elif(entry[i]==111):   #111
                Matrix[i]=rule[7]
                
        print(Matrix)
        
   
```

```python
def main():
    #Bynary number from 0 to 255
    entry=50
    L=20
    time=100
    
    binary='{0:08b}'.format(entry)  #Converts decimal into binary
    Automata(L,list(binary),time)
```
